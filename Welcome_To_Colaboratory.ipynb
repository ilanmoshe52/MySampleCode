```
import random
import matplotlib.pyplot as plt
```

```
n = 2000
```

```
x = [random.normalvariate(0,1) for i in range (n)]
```

```
plt.plot(x)
```

```
def turning(s):
    Tplist = [((s[i]> s[i-1]) and (s[i]> s[i+1])) or ((s[i]< s[i-1]) and (s[i]< s[i+1])) for i in range (1,len(s)-1) ]
    n = len(s)
    mt = 2/3*(n-2)
    vt = (16*n-29)/90
    if abs(sum(Tplist) - mt)/vt**0.5>1.96:
       print('reject')
    else:
       print('accept')
    print(sum(Tplist))
```

```
turning(x)
```

```
y = [x[i]+x[i-1]*0.3 for i in range (1,n)]
```

```
turning(y)
```
